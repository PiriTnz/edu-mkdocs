```python
import pandas as pd

# Load the dataset
data = pd.read_csv('cars_synth_clean.csv')
```

```python
from sklearn.model_selection import train_test_split

# Split the dataset into train and test sets
train_set, test_set = train_test_split(data, test_size=0.33, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression

# Selecting features and target variable
X_train = train_set[['length']]
y_train = train_set['price']

# Initialize and train the model
model = LinearRegression()
model.fit(X_train, y_train)

# Predict on the train set
train_predictions = model.predict(X_train)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import numpy as np

# Compute scores on the train set
mse = mean_squared_error(y_train, train_predictions)
rmse = np.sqrt(mse)
mae = mean_absolute_error(y_train, train_predictions)
r2 = r2_score(y_train, train_predictions)

print(f"Train MSE: {mse}, RMSE: {rmse}, MAE: {mae}, R²: {r2}")
```

```python
X_test = test_set[['length']]
y_test = test_set['price']

# Predict on the test set
test_predictions = model.predict(X_test)

# Compute scores on the test set
mse_test = mean_squared_error(y_test, test_predictions)
rmse_test = np.sqrt(mse_test)
mae_test = mean_absolute_error(y_test, test_predictions)
r2_test = r2_score(y_test, test_predictions)

print(f"Test MSE: {mse_test}, RMSE: {rmse_test}, MAE: {mae_test}, R²: {r2_test}")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Scatter plot of true vs predicted
plt.figure(figsize=(8, 6))
sns.scatterplot(x=y_test, y=test_predictions)
plt.xlabel("True Price")
plt.ylabel("Predicted Price")
plt.title("True Price vs Predicted Price")
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], color='red')  # Diagonal line
plt.show()
```
